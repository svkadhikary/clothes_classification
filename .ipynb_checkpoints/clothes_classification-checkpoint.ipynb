```python
# lets look at a sample image
img = image.load_img("./clothing-dataset-small/train/outwear/00149032-3dd6-426e-9bc0-d53032536a42.jpg")
imshow(img)
```

```python
# hard setting the image shape and batch size
image_size = (224, 224)
batch_size = 20
```

```python
print(tf.__version__)
```

#### Use image_dataset_from_directory from tf.keras.utils to load the images and labels inside our dataset. The hierarchy of the images in their respective labels folder are pre-sorted.

```python
# path to training images
train_dir = pathlib.Path("./clothing-dataset-small/train/")
with tf.device('/cpu:0'):
    train_ds = tf.keras.utils.image_dataset_from_directory(
        train_dir,
        seed=25,
        image_size=image_size,
        batch_size=batch_size,
    )
```

```python
# path to validation images
val_dir = pathlib.Path("./clothing-dataset-small/validation/")
with tf.device('/cpu:0'):
    val_ds = tf.keras.utils.image_dataset_from_directory(
        val_dir,
        seed=25,
        image_size=image_size,
        batch_size=batch_size
    )
```

```python
# path to test set of images
test_dir = pathlib.Path("./clothing-dataset-small/test/")
with tf.device('/cpu:0'):
    test_ds = tf.keras.utils.image_dataset_from_directory(
        test_dir,
        seed=25,
        image_size=image_size,
        batch_size=batch_size,
    )
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python
# Lets check some images
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
      for i in range(9):
        ax = plt.subplot(3, 3, i+1)
        plt.imshow(images[i].numpy().astype('uint8'))
        plt.title(class_names[labels[i]])
        plt.axis("off")

plt.show()
```

```python
# image and label batch shapes
for image_batch, label_batch in train_ds:
    print(image_batch.shape)
    print(label_batch.shape)
    break
```

#### Lets declare our rescaling layer which will convert the pixel values of the images between -1 and 1, since our pre-trained model is trained in that constraint. Also convert the labels to one-hot values.

```python
# Rescaling layer
normalization_layer = tf.keras.layers.Rescaling(scale=1./127.5, offset=-1)
```

```python
# One-hot encoding
def convert_to_one_hot(y, d):
      return tf.one_hot(y, d)
```

```python
# A simple tweak to boost performance, if the retrieval of data lags
AUTOTUNE = tf.data.AUTOTUNE
with tf.device('/cpu:0'):
    train_dataset = train_ds.prefetch(buffer_size=AUTOTUNE)
    validation_dataset = val_ds.prefetch(buffer_size=AUTOTUNE)
    test_dataset = test_ds.prefetch(buffer_size=AUTOTUNE)
```

```python
# data augmentation function if overfitting happens
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.RandomFlip('horizontal'),
  tf.keras.layers.RandomRotation(0.2),
])
```

```python
# preprocessing of data
with tf.device('/cpu:0'):
    train_dataset = train_dataset.map(lambda x, y: (normalization_layer(x), convert_to_one_hot(y, 10)))
    validation_dataset = validation_dataset.map(lambda x, y: (normalization_layer(x), convert_to_one_hot(y, 10)))
    test_dataset = test_dataset.map(lambda x, y: (normalization_layer(x), convert_to_one_hot(y, 10)))
```

```python
# a look at how the data augmentation will convert 
for image, _ in train_dataset.take(1):    
    plt.figure(figsize=(10, 10))
    first_image = image[0]
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        augmented_image = data_augmentation(tf.expand_dims(first_image, 0))
        plt.imshow(augmented_image[0])
        plt.axis('off')
```

#### As stated above lets use mobilenetv2 pretrained model for classification.
##### At first excluding the top classifier layer and add our own relu layer and then classifier layer with 10 units since there are only 10 classes.

```python
# initialize mobilenetv2 as base model
tf.keras.backend.clear_session()
tf.debugging.set_log_device_placement(True)

new_base_model = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3),
                                               include_top=False,
                                               weights='imagenet')
new_base_model.trainable = False
```

```python
# computation graph

x = new_base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dropout(0.2)(x)
predictions = tf.keras.layers.Dense(10, activation='softmax')(x)

new_model = tf.keras.models.Model(inputs=new_base_model.input, outputs=predictions)

new_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# lets run the model for 10 epochs to see how it performs
initial_epochs = 10
history = new_model.fit(train_dataset,
                            validation_data=validation_dataset,
                            epochs=initial_epochs)
```

```python
# Lets check the accuracy on test dataset
loss, accuracy = new_model.evaluate(test_dataset)
print("accuracy:", str(accuracy))
```

#### Quite good accuracy in 10 epochs only. Lets see if unfreezing some bottom layers will input better accuuracy on the test set or not. Unfreezing the bottom layers first can allow the model to learn more general features from the data, which can be helpful since the dataset is small and may not contain enough examples to fully train the top layers of the model.

```python
# unfreezeing first 15 layers
tf.keras.backend.clear_session()

for layer in new_base_model.layers[:50]:
    layer.trainable = True

new_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

len(new_model.trainable_variables)
```

```python
# lets also set a callback for saving the best model with and highest val accuracy.
# Another callback for early stopping monitoring the val loss
# And another for learningrate decay when the loss and accuracy hits a plateau
mcp_callback = tf.keras.callbacks.ModelCheckpoint(filepath='weights.{epoch:02d}-{val_accuracy:.2f}-{val_loss:.2f}.h5', save_best_only=True, monitor='val_loss', mode='min')
earlyStopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, verbose=0, mode='min', restore_best_weights=True)
reduce_lr_loss = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=2, verbose=1, min_delta=1e-4, mode='min')
```

```python
# train the unfreezed model

history_fine = new_model.fit(train_dataset,
                            validation_data=validation_dataset,
                            epochs=initial_epochs+10,
                            initial_epoch=history.epoch[-1],
                            callbacks=[mcp_callback, earlyStopping, reduce_lr_loss])
```

```python
tf.keras.backend.clear_session()
```

```python
# Accuracy on the test set
loss, accuracy = new_model.evaluate(test_dataset)
print("accuracy:", str(accuracy))
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
acc += history_fine.history['accuracy']
val_acc += history_fine.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']
loss += history_fine.history['loss']
val_loss += history_fine.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.plot([initial_epochs-1,initial_epochs-1],
          plt.ylim(), label='Start Fine Tuning')
plt.legend(loc='lower right')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([0,1.0])
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.plot([initial_epochs-1,initial_epochs-1],
         plt.ylim(), label='Start Fine Tuning')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,1.0])
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

### Not much improvement over the previous results. Maybe due to the low number of data the model is not learning enough.

#### Lets look at some pictures from the test set and their predictions by the model

```python
image_batch, label_batch = next(iter(test_dataset))
predictions = new_model.predict_on_batch(image_batch)
```

```python
predictions = np.where(predictions < 0.5, 0, 1)
predictions = tf.argmax(predictions, axis=1)
label_batch = np.argmax(label_batch, axis=1)

print('Predictions:\n', predictions.numpy())
print('Labels:\n', label_batch)

plt.figure(figsize=(15, 15))
for i in range(20):
    if label_batch[i] == predictions[i]:
        color = 'green'
    else:
        color = 'red'
    ax = plt.subplot(5, 4, i + 1)
    plt.imshow(image_batch[i].numpy())
    plt.title(class_names[predictions[i]], color=color)
    plt.axis("off")
plt.show()
```

### Quite good predictions. Lets also check with self-captured images

```python
self_cap_dir = "./self-captured"
i = 1
plt.figure(figsize=(10, 10))
for file in os.listdir(self_cap_dir):
    filepath = os.path.join(self_cap_dir, file)
    # preprocess the image
    img = tf.keras.preprocessing.image.load_img(filepath)
    img = tf.keras.preprocessing.image.img_to_array(img)
    img = tf.image.resize(img, (224, 224))
    img = np.expand_dims(img, 0)
    # resccale the image to fit the model preferred input
    img_rescaled = normalization_layer(img)
    # prediction
    pred = np.argmax(new_model(img_rescaled))
    # prints the image
    plt.subplot(6, 4, i)
    plt.imshow(img[0]/255)
    plt.title("Predicted: " + class_names[pred])
    plt.axis("off")
    i += 1
plt.show()
```
